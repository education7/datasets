```python
url = "https://raw.githubusercontent.com/education7/datasets/main/bank.csv"
```

```python
import pandas as pd
from sklearn.pipeline import make_pipeline
df = pd.read_csv(url )
```

```python
import numpy as np
df = df.replace("?",np.nan)
```

```python
def cat_var(L):
  for i in L :
    a = dict(df[i].value_counts())
    b = list(a.keys())
    df[i] = df[i].fillna(b[0])
```

```python
def num_mean(L):
  for i in L :
    df[i] = df[i].replace(np.NAN,df[i].mean())

def num_median(L):
  for i in L :
    df[i] = df[i].replace(np.nan,df[i].mean())
```

```python
L = ['job','marital','education','default','housing','loan','contact','month','poutcome']
cat_var(L)
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')
df['day'] = pd.to_numeric(df['day'], errors='coerce')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
df['campaign'] = pd.to_numeric(df['campaign'], errors='coerce')
df['previous'] = pd.to_numeric(df['previous'], errors='coerce')
df['pdays'] = pd.to_numeric(df['pdays'], errors='coerce')
```
